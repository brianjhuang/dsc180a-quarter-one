```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = json.load(open('../runs/all_runs/run_1.json'))
correct_labels = []

with open('../data/raw_data/video_labels.txt') as f:
    for label in f:
        correct_labels.append(label.strip('\n'))
```

```python
df = pd.DataFrame(data['items'])
df['actual'] = correct_labels
df['correct'] = df['actual'] == df['prediction']
```

```python
df
```

```python
labels = ['actual', 'predictions']
science = [(df['actual'] == 'science').sum(), (df['prediction'] == 'science').sum()]
pseudo = [(df['actual'] == 'pseudoscience').sum(), (df['prediction'] == 'pseudoscience').sum()]

axis = np.arange(len(labels))

plt.bar(axis - 0.2, science, 0.4, label = 'Science')
plt.bar(axis + 0.2, pseudo, 0.4, label = 'Pseudoscience')

plt.xticks(axis, labels)
plt.ylabel("Number of Videos")
plt.title("Actual vs Prediction")
plt.legend()
plt.show()
```
